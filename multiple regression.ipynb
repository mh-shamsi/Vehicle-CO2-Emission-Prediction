```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab as pl
```

```python
df=pd.read_csv("FuelConsumption.csv")
df.head()
```

```python
df.describe()
```

```python
cdf=df[["ENGINESIZE",'CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
```

```python
plt.plot(cdf['ENGINESIZE'],cdf['CO2EMISSIONS'],"bo")

plt.show()
```

```python
plt.plot(cdf['CYLINDERS'],cdf['CO2EMISSIONS'],"bo")
```

```python
plt.plot(cdf['FUELCONSUMPTION_COMB'],cdf['CO2EMISSIONS'],"bo")
```

```python
mask=np.random.rand(len(cdf)) < 0.8
train=cdf[mask]
test=cdf[~mask]
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
trainx=np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
trainy=np.asanyarray(train[['CO2EMISSIONS']])
regressor.fit(trainx,trainy)
print(regressor.intercept_)
print(regressor.coef_)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
testx=np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
testy=np.asanyarray(test[['CO2EMISSIONS']])
y_hat=regressor.predict(testx)
print("r2 :%.2f" % r2_score(testy,y_hat))
print("mse:%.2f" % mean_squared_error(testy,y_hat))
```
